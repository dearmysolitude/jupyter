### 분류 모델 생성

```python
from keras import utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np

from keras.callbacks import EarlyStopping
from keras.callbacks import TensorBoard
import numpy as np
import sys
import matplotlib.pyplot as plt

# 1. 데이터셋 생성
## 훈련셋과 테스트셋 가져오기
(x_train, y_train), (x_test, y_test) = mnist.load_data()
## 전처리: 2차원 리스트를 1차원 리스트로 / 데이터의 값들을 0~1사이 값으로 변환(표준화)
x_train = x_train.reshape(60000, 784).astype('float32')/255.0
x_test = x_test.reshape(10000, 784).astype('float32')/255.0
## class 데이터 원핫 인코딩
# y_train = np_utils.to_categorical(y_train)
# y_test = np_utils.to_categorical(y_test)
y_train = utils.to_categorical(y_train)
y_test = utils.to_categorical(y_test)

# 2. 모델 정의 구성
model = Sequential()
model.add(Dense(units=128, input_dim=784, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=10, activation='softmax'))

# 3. 모델 학습과정 설정
# model.compile(loss='categorical_croosentropy', optimizer='sgd', metrics=['accuracy'])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. 모델 학습시키기
model.fit(x_train, y_train, validation_split=0.2, epochs=10, batch_size=500, verbose=1)

# 5. 모델 평가하기
score = model.evaluate(x_test, y_test, batch_size=32)
print("")
print("test score: ", score)

# 모델 저장
model.save("mnist_my_model.h5")
```

## CNN
- Convolutional Neural Network, 합성곱 신경망
- 인간의 시신경 구조를 모방. 이미지 인식에 많이 사용한다.
- [특징을 추출하는 feature extraction] + [feature extraction을 통과한 이후에 결과값을 도추랗는 classification] 으로 구성
- Feature extraction: Convolution layer와 Pooling layer가 섞여있다.
- Classification: fully-connected layer

```python
import keras
from keras.utils import np_utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
import numpy as np

img_rows = 28
img_cols = 28

(x_train, y_train), (x_test, y_test) = mnist.load_data()

input_shape = (img_rows, img_cols, 1)
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)

x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

num_classes = 10
y_train = keras.utils.np_utils.to_categorical(y_train, num_classes)
y_test = keras.utils.np_utils.to_categorical(y_test, num_classes)

import time
start = time.time()

model = Sequential()
model.add(Conv2D(32, kernel_size=(5, 5), strides=(1,1), 
                 padding='same', activation='relu', input_shape=input_shape))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(64, (2, 2), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

hist = model.fit(x_train, y_train, 
                 batch_size=128, epochs=12, verbose=1, validation_split=0.2)

score = model.evaluate(x_test, y_test, verbose=0)
print('Test score: ', score)
print("Proc-time: ", time.time() - start)

model.save('mnist_my_cnn_model.h5')
```

## 활용: 글자 이미지가 궁서체인지 아닌지 판단하는 모델 만들기
### 서체 이미지 만들기

```python
import os
from PIL import Image, ImageDraw, ImageFont
import random
# 이미지 생성 폴더 경로
gungseo_path = './font_images/gungseo_images'
non_gungseo_path = './font_images/non_gungseo_images'
os.makedirs(gungseo_path, exist_ok=True)
os.makedirs(non_gungseo_path, exist_ok=True)
# 궁서체 및 다른 폰트 경로
gungseo_font_path = 'font/gungseo.ttc'  # 궁서체 폰트 파일 경로
other_font_path = 'font/malgun.ttf'  # 다른 폰트 파일 경로
# 폰트 크기 설정
font_size = 20
# 한글 자모 리스트
chosung = ["ㄱ", "ㄲ", "ㄴ", "ㄷ", "ㄸ", "ㄹ", "ㅁ", "ㅂ", "ㅃ", "ㅅ", "ㅆ", "ㅇ", "ㅈ", "ㅉ", "ㅊ", "ㅋ", "ㅌ", "ㅍ", "ㅎ"]
jungsung = ["ㅏ", "ㅐ", "ㅑ", "ㅒ", "ㅓ", "ㅔ", "ㅕ", "ㅖ", "ㅗ", "ㅘ", "ㅙ", "ㅚ", "ㅛ", "ㅜ", "ㅝ", "ㅞ", "ㅟ", "ㅠ", "ㅡ", "ㅢ", "ㅣ"]
jongsung = ["", "ㄱ", "ㄲ", "ㄳ", "ㄴ", "ㄵ", "ㄶ", "ㄷ", "ㄹ", "ㄺ", "ㄻ", "ㄼ", "ㄽ", "ㄾ", "ㄿ", "ㅀ", "ㅁ", "ㅂ", "ㅄ", "ㅅ", "ㅆ", "ㅇ", "ㅈ", "ㅊ", "ㅋ", "ㅌ", "ㅍ", "ㅎ"]
# 유니크한 한글 글자 생성
unique_chars = set()
while len(unique_chars) < 6000:
    chosung_char = random.choice(chosung)
    jungsung_char = random.choice(jungsung)
    jongsung_char = random.choice(jongsung)
    unique_char = chr(0xAC00 + (chosung.index(chosung_char) * 21 * 28) + (jungsung.index(jungsung_char) * 28) + jongsung.index(jongsung_char))
    unique_chars.add(unique_char)
unique_chars = list(unique_chars)
# 이미지 생성 함수
def create_image(text, font_path, output_path, index):
    font = ImageFont.truetype(font_path, font_size)
    image = Image.new('RGB', (28, 28), color=(255, 255, 255))
    draw = ImageDraw.Draw(image)
    text_bbox = draw.textbbox((0, 0), text, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]
    position = ((28 - text_width) / 2, (28 - text_height) / 2)
    draw.text(position, text, font=font, fill=(0, 0, 0))
    image.save(os.path.join(output_path, f'image_{index}.png'))
# 6000개 이상의 이미지 생성
for i in range(6000):
    text = unique_chars[i]
    create_image(text, gungseo_font_path, gungseo_path, i)
    create_image(text, other_font_path, non_gungseo_path, i)
```

```python
from keras.datasets import reuters
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
from keras_preprocessing import sequence
from keras.utils import np_utils

import numpy
import tensorflow as tf
import matplotlib.pyplot as plt

# 데이터 불러오기
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=1000, test_split=0.2)

# 전처리
x_train = sequence.pad_sequences(x_train, maxlen=100)
x_test = sequence.pad_sequences(x_test, maxlen=100)
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)

model = Sequential()
model.add(Embedding(1000, 200))
model.add(LSTM(100, activation='tanh'))
model.add(Dense(46,  activation='softmax'))

# 모델 설정, 학습
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
history = model.fit(x_train, y_train, batch_size = 100, epochs=50, validation_data=(x_test, y_test))

# 그래프 그리기
y_vloss = history.history['val_loss']
y_loss = history.history['loss']
x_len = numpy.arange(len(y_loss))
plt.plot(x_len, y_vloss, c='r')
plt.plot(x_len, y_loss, c='b')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
