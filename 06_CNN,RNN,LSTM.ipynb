### 이미지 출력해보기

```python
from keras.utils import np_utils
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

get_pt = 22 # 임의로 불러올 데이터

(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(" ", x_train.shape)

# 이미지 출력
plt.figure(figsize=(5,5))
image = np.reshape(x_train[get_pt], [28, 28])
plt.imshow(image, cmap='Greys')
plt.show()

# 실제값
print("################################################")
x_train_df = pd.DataFrame(x_train[get_pt])
pd.set_option("display.max_rows", 150)
pd.set_option("display.max_columns", 150)
print("target: ", y_train[get_pt])
x_train_df
```

### 분류 모델 생성

```python
from keras import utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np

from keras.callbacks import EarlyStopping
from keras.callbacks import TensorBoard
import numpy as np
import sys
import matplotlib.pyplot as plt

# 1. 데이터셋 생성
## 훈련셋과 테스트셋 가져오기
(x_train, y_train), (x_test, y_test) = mnist.load_data()
## 전처리: 2차원 리스트를 1차원 리스트로 / 데이터의 값들을 0~1사이 값으로 변환(표준화)
x_train = x_train.reshape(60000, 784).astype('float32')/255.0
x_test = x_test.reshape(10000, 784).astype('float32')/255.0
## class 데이터 원핫 인코딩
y_train = utils.to_categorical(y_train)
y_test = utils.to_categorical(y_test)

# 2. 모델 정의 구성
model = Sequential()
model.add(Dense(units=128, input_dim=784, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=10, activation='softmax'))

# 3. 모델 학습과정 설정
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 4. 모델 학습시키기
model.fit(x_train, y_train, validation_split=0.2, epochs=10, batch_size=500, verbose=1)

# 5. 모델 평가하기
score = model.evaluate(x_test, y_test, batch_size=32)
print("")
print("test score: ", score)

# 모델 저장
model.save("mnist_my_model.h5")
```

### 생성한 모델로 테스트해보기

```python
import numpy as np
# 6. 모델로 예측하기
want_test_size = 500 # 테스트셋 중 고를 랜덤 데이터는 500개
xhat_idx = np.random.choice(x_test.shape[0], want_test_size)

xhat = x_test[xhat_idx]
print("**********************")

# tf 2.6 이전
# yhat = model.predict_classes(xhat) # 5개의 데이터 예측시킴
# tf 2.6 부터
y_prob = model.predict(xhat, verbose=0)
yhat = y_prob.argmax(axis=-1)
####################################################
ok_cnt = 0 # 맞춘 갯수
for i in range(want_test_size):
    if np.argmax(y_test[xhat_idx[i]]) == yhat[i]:
        ok_cnt = ok_cnt + 1
        print("OK 실제: ", np.argmax(y_test[xhat_idx[i]]), "예측: ", yhat[i])
    else:
        print("NOK ==> 실제: ", np.argmax(y_test[xhat_idx[i]]), "예측: ", yhat[i])
####################################################
print("총 ", want_test_size, "개 중, ", ok_cnt, "개 예측 성공 / 정확도: %.2f%%" %((ok_cnt/want_test_size) * 100))
```
