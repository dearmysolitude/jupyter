```python
from keras import utils
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.callbacks import EarlyStopping
from keras.callbacks import TensorBoard

import numpy as np
import sys
from time import time

# 1. 데이터셋 생성
## 훈련셋과 시험셋 불러오기
(x_train, y_train), (x_test, y_test) = mnist.load_data()

## 2차원을 1차원으로 바꾸고, 255를 나누어 1보다 작은 소수점으로 만든다.
x_train = x_train.reshape(60000, 784).astype('float32')/255.0
x_test = x_test.reshape(10000, 784).astype('float32')/255.0

## class 데이터 원핫 인코딩 처리: tensor flow 버전에 따라: np_utils -> utils
y_train = utils.to_categorical(y_train)
y_test = utils.to_categorical(y_test)

## 훈련셋과 검증셋 분리
x_val = x_train[42000:]
x_train = x_train[:42000]
y_val = y_train[42000:]
y_train = y_train[:42000]

# 2. 모델 정의 구성
model = Sequential()
model.add(Dense(units=36, input_dim=28*28, activation='relu'))
model.add(Dense(units=10, activation='softmax'))

# 3. 모델 학습과정 설정
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

s_t = time()
print("start: ", s_t)

# 4. 모델 학습시키기
model.fit(x_train, y_train, epochs=100, batch_size=50, validation_data=(x_val, y_val), verbose=0)
e_t = time()
print("end: ", e_t, "time: ", e_t-s_t)

# 5. 모델 평가하기
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=32)
print("")
print("loss_and_metrics: "+ str(loss_and_metrics))

# 6. 모델 사용하여 예측해보기
xhat_idx = np.random.choice(x_test.shape[0], 5) # 데이터셋 랜덤 설정

xhat = x_test[xhat_idx]

# tensorflow 2.6 이전
# yhat = model.predict_classes(xhat)

# 2.6 이후
y_prob = model.predict(xhat, verbose=0)
yhat = y_prob.argmax(axis=-1)

for i in range(5):
    print("True: " + str(np.argmax(y_test[xhat_idx[i]]))+", Predict: "+str(yhat[i]))
# 5개 출력
```
